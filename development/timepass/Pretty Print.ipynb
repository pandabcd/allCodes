```python
import os
# your code goes here
def get_new_line(string):
    return  os.linesep

def get_emoji_length(emoji_name):
    '''Takes a emoji name as input and returns its approximate length in units of a space'''
    
    emoji_length = {}
    emoji_length[':angry:'] = 6
    emoji_length[':sunglasses:'] = 6
    emoji_length[':smile:'] = 6
    emoji_length[':woman-shrugging:'] = 6
    
    
    if emoji_name not in emoji_length:
        return 6
    
    return emoji_length[emoji_name]


def get_emoji_row(digital_row, emoji_name):
    '''Helper function for building emoji matrix(builds row by row)'''
    emoji_row = ""
    emoji_length = get_emoji_length(emoji_name)
    
    for i in xrange(0,len(digital_row[0]), 2):
        if digital_row[0][i] == '-':
            emoji_row += " "*emoji_length
        else:
            emoji_row += emoji_name
        

#     for elem in digital_row:
#         if elem == '-':
#             emoji_row += " "*emoji_length
#         else:
#             emoji_row += emoji_name
            
    emoji_row += get_new_line(emoji_name)
    return emoji_row
            

def get_emoji_matrix(digital_letter_shape, emoji_name):
    '''Takes 2-d matrix of a letter and emoji name and constructs emoji matrix for the letter'''
    emoji_matrix = ""
    
    for digital_row in digital_letter_shape:
        emoji_matrix += get_emoji_row(digital_row, emoji_name)
    
    return emoji_matrix

           
def add_magical_character(shape_word, shape_char):
    for i, row in enumerate(shape_word):
        shape_word[i][0] += ' - - '
    
    for i, row in enumerate(shape_word):
        shape_word[i][0] += shape_char[i][0]
    
    return shape_word
    
def get_magical_character(letter, emoji_name):
    shape_letter = get_shape(ord(letter) - ord('a'))
    return get_emoji_matrix(shape_letter, emoji_name)

def get_magical_string(word, emoji_name):
    final_magic_string = ""
    
    shape_word = get_shape(ord(word[0]) - ord('a'))
    
#     print shape_word
    
    for i in xrange(1, len(word)):
        shape_char = get_shape( ord(word[i]) - ord('a'))
        shape_word = add_magical_character(shape_word, shape_char )
    
#     for row in shape_word:
#         print row
    
    return get_emoji_matrix(shape_word, emoji_name)
        
        
        
        
#         for i, digital_row in enumerate(string_matrix):
# #         print string_matrix[i]
#         string_matrix[i].extend(['-', '-'])
# #         print string_matrix[i]
# #         print
#     for i, digital_row in enumerate(string_matrix):
#         string_matrix[i].extend(letter1[i])
    
#     for i, digital_row in enumerate(string_matrix):
#         print digital_row
```
